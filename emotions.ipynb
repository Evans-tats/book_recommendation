```python
import pandas as pd

books = pd.read_csv("./books_with_categories")
```

```python
books
```

```python
books['simple_categories'] = books['simple_categories'].str.replace('non_fiction', 'Nonfiction')
books['simple_categories'] = books['simple_categories'].str.replace('fiction', 'Fiction')
books.head(5)
```

```python
from transformers import pipeline
classifier = pipeline(
    "text-classification",
    model = "j-hartmann/emotion-english-distilroberta-base",
    return_all_scores = True
)
classifier("heyy mum")
```

```python
classifier(books["description"][0])
```

```python
emotion_labels = ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surpise']
isbn = []
emotion_scores = {label: [] for label in emotion_labels }

def calculate_max_emotion_score(predictions):
  per_emotion_scores = {label: [] for label in emotion_labels }
  for prediction in predictions:
    sorted_predictions = sorted(prediction, key=lambda x: x["label"])
    for index, label in enumerate(emotion_labels):
      per_emotion_scores[label].append(sorted_predictions[index]["score"])
  return {label: max(scores) for label, scores in per_emotion_scores.items()}
```

```python
for i in range(10):
  isbn.append(books["isbn13"][i])
  sentences = books["description"][i].split(".")
  predictions = classifier(sentences)
  max_score = calculate_max_emotion_score(predictions)
  for label in emotion_labels:
    emotion_scores[label].append(max_score[label])
```

```python
emotion_scores
```

```python
from tqdm import tqdm

emotion_labels = ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surpise']
isbn = []
emotion_scores = {label: [] for label in emotion_labels }

for i in tqdm(range(len(books))):
  isbn.append(books["isbn13"][i])
  sentences = books["description"][i].split(".")
  predictions = classifier(sentences)
  max_score = calculate_max_emotion_score(predictions)
  for label in emotion_labels:
    emotion_scores[label].append(max_score[label])
```

```python
emotion_df = pd.DataFrame(emotion_scores)
emotion_df
```

```python
emotion_df["isbn13"] = isbn
emotion_df
```

```python
books_with_emotion = pd.merge(books, emotion_df, on="isbn13")
```
